```python
## Exercise one 
import pandas as pd
data_old = pd.read_csv("../data/Police_Department_Incident_Reports__Historical_2003_to_May_2018_20250225.csv")
data_new = pd.read_csv("../data/Police_Department_Incident_Reports__2018_to_Present_20250225.csv")

# cols in 1['PdId', 'IncidntNum', 'Incident Code', 'Category', 'Descript', 'DayOfWeek', 'Date', 'Time', 'PdDistrict', 'Resolution', 'Address', 'X', 'Y', 'location', 'SF Find Neighborhoods 2 2', 'Current Police Districts 2 2', 'Current Supervisor Districts 2 2', 'Analysis Neighborhoods 2 2', 'DELETE - Fire Prevention Districts 2 2', 'DELETE - Police Districts 2 2', 'DELETE - Supervisor Districts 2 2', 'DELETE - Zip Codes 2 2', 'DELETE - Neighborhoods 2 2', 'DELETE - 2017 Fix It Zones 2 2', 'Civic Center Harm Reduction Project Boundary 2 2', 'Fix It Zones as of 2017-11-06  2 2', 'DELETE - HSOC Zones 2 2', 'Fix It Zones as of 2018-02-07 2 2', 'CBD, BID and GBD Boundaries as of 2017 2 2', 'Areas of Vulnerability, 2016 2 2', 'Central Market/Tenderloin Boundary 2 2', 'Central Market/Tenderloin Boundary Polygon - Updated 2 2', 'HSOC Zones as of 2018-06-05 2 2', 'OWED Public Spaces 2 2', 'Neighborhoods 2']
columns_to_keep1 = ['Date','Time','Category','DayOfWeek','X','Y','PdDistrict'] #Adjust the features you want to keep
data_old = data_old[columns_to_keep1]
# cols in 2 ['Incident Datetime', 'Incident Date', 'Incident Time', 'Incident Year', 'Incident Day of Week', 'Report Datetime', 'Row ID', 'Incident ID', 'Incident Number', 'CAD Number', 'Report Type Code', 'Report Type Description', 'Filed Online', 'Incident Code', 'Incident Category', 'Incident Subcategory', 'Incident Description', 'Resolution', 'Intersection', 'CNN', 'Police District', 'Analysis Neighborhood', 'Supervisor District', 'Supervisor District 2012', 'Latitude', 'Longitude', 'Point', 'Neighborhoods', 'ESNCAG - Boundary File', 'Central Market/Tenderloin Boundary Polygon - Updated', 'Civic Center Harm Reduction Project Boundary', 'HSOC Zones as of 2018-06-05', 'Invest In Neighborhoods (IIN) Areas', 'Current Supervisor Districts', 'Current Police Districts']
columns_to_keep2 = ['Incident Datetime','Incident Day of Week','Incident Category'
                   ,'Latitude','Longitude','Police District']
data_new = data_new[columns_to_keep2]

category_mapping = {
    # Focus crimes
    'Robbery': 'ROBBERY',
    'Burglary': 'BURGLARY',
    'Assault': 'ASSAULT',
    'Drug Violation': 'DRUG/NARCOTIC',
    'Drug Offense': 'DRUG/NARCOTIC',
    'Larceny Theft': 'LARCENY/THEFT',
    'Vandalism': 'VANDALISM',
    'Malicious Mischief': 'VANDALISM',  
    'Motor Vehicle Theft': 'VEHICLE THEFT',
    'Motor Vehicle Theft?': 'VEHICLE THEFT',
    'Stolen Property': 'STOLEN PROPERTY',
    'Disorderly Conduct': 'DISORDERLY CONDUCT',
    'Weapons Carrying Etc': 'WEAPON LAWS',
    'Weapons Offense': 'WEAPON LAWS',
    'Weapons Offence': 'WEAPON LAWS',
    'Prostitution': 'PROSTITUTION',
    # Additional mappings for edge cases
    'Traffic Violation Arrest': 'DRIVING UNDER THE INFLUENCE',  # If DUI is included here 
    'Suspicious Occ': 'TRESPASS',
    'Suspicious': 'TRESPASS',
    'Liquor Laws': 'DRUNKENNESS'  # If liquor law violations include public drunkenness 
}
focus_crimes = [
    'WEAPON LAWS', 'PROSTITUTION', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUG/NARCOTIC', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY','OTHER OFFENSES']

data_new['Incident Category'] = data_new['Incident Category'].replace(category_mapping)


df1_filtered = data_old.copy()
df2_filtered = data_new.copy()


df2_filtered['Incident Category'] = df2_filtered['Incident Category'].apply(lambda x: x if x in focus_crimes else 'OTHER OFFENSES')
df1_filtered['Category'] = df1_filtered['Category'].apply(lambda x: x if x in focus_crimes else 'OTHER OFFENSES')

df1_filtered = df1_filtered[df1_filtered['Category'].isin(focus_crimes)] # Filter the rows for the non-focused crime categories
df2_filtered = df2_filtered[df2_filtered['Incident Category'].isin(focus_crimes)] # Filter the rows for the non-focused crime categories


# Parsing the date and time
df1_filtered = pd.DataFrame(df1_filtered)

# Convert "Date" column to datetime format
df1_filtered["Date"] = pd.to_datetime(df1_filtered["Date"], format="%m/%d/%Y")

# Create new columns from "Date"
df1_filtered["DayOfMonth"] = df1_filtered["Date"].dt.day
df1_filtered["Month"] = df1_filtered["Date"].dt.strftime("%B")  # Month name
df1_filtered["Year"] = df1_filtered["Date"].dt.year

# Extract the hour from the "Time" column to create "TimeOfDay"
df1_filtered["TimeOfDay"] = pd.to_datetime(df1_filtered["Time"], format="%H:%M").dt.hour
df1_filtered = df1_filtered.rename(columns={"X": "Longitude", "Y": "Latitude"})
df1_filtered = df1_filtered.drop(columns=["Date", "Time"])
# Parsing the date and time
df2_filtered = pd.DataFrame(df2_filtered)

# Convert "Incident Datetime" to a datetime format
df2_filtered["Incident Datetime"] = pd.to_datetime(df2_filtered["Incident Datetime"], format="%Y/%m/%d %I:%M:%S %p")

# Create new columns from "Incident Datetime"
df2_filtered["DayOfMonth"] = df2_filtered["Incident Datetime"].dt.day
df2_filtered["Month"] = df2_filtered["Incident Datetime"].dt.strftime("%B")  # Month name
df2_filtered["Year"] = df2_filtered["Incident Datetime"].dt.year
df2_filtered["TimeOfDay"] = df2_filtered["Incident Datetime"].dt.hour
df2_filtered = df2_filtered.rename(columns={"Incident Category": "Category", "Incident Day of Week": "DayOfWeek"})
df2_filtered = df2_filtered.drop(columns=["Incident Datetime"])
df2_filtered = df2_filtered.rename(columns={"Police District": "PdDistrict"})

#drop na ["Latitude", "Longitude","Category","PdDistrict"])
df1_filtered = df1_filtered.dropna(subset=["Latitude", "Longitude","Category","PdDistrict"]) # Get rid of missing values
df2_filtered = df2_filtered.dropna(subset=["Latitude", "Longitude","Category","PdDistrict"]) # Get rid of missing values



df2_filtered = df2_filtered[df2_filtered['PdDistrict'] != 'Out of SF'] # this value is only available in 2nd dataset so we remove it


pd_mapping = {
    # PdDistrict text values should match in both datasets
    'Bayview': 'BAYVIEW',
    'Mission': 'MISSION',
    'Northern': 'NORTHERN',
    'Central': 'CENTRAL',
    'Richmond': 'RICHMOND',
    'Ingleside': 'INGLESIDE',
    'Southern': 'SOUTHERN',
    'Park': 'PARK',  
    'Taraval': 'TARAVAL',
    'Tenderloin': 'TENDERLOIN',
}

# Apply the mapping to df2's "PdDistrict"
df2_filtered['PdDistrict'] = df2_filtered['PdDistrict'].replace(pd_mapping)

# Ensure that both DataFrames have the same columns in the same order
df1_filtered = df1_filtered[['Category', 'PdDistrict', 'Longitude', 'Latitude', 'TimeOfDay', 'DayOfWeek', 'DayOfMonth', 'Month', 'Year']]
df2_filtered = df2_filtered[['Category', 'PdDistrict', 'Longitude', 'Latitude', 'TimeOfDay', 'DayOfWeek', 'DayOfMonth', 'Month', 'Year']]
# Drop rows where Year is 2018
df1_filtered = df1_filtered[df1_filtered['Year'] != 2018]
df2_filtered = df2_filtered[df2_filtered['Year'] != 2025]
merged_data = pd.concat([df1_filtered, df2_filtered], axis=0, ignore_index=True) # Perform concat


# Save the merged data to a CSV file
merged_data.to_csv("../data/merged_data.csv", index=False)
```

```python
import pandas as pd
from matplotlib import pyplot as plt
merged_data = pd.read_csv("../data/merged_data.csv")
print("Number of crimes", merged_data.shape[0])
print("Cataegories of crimes", merged_data['Category'].unique())
print(merged_data["Category"].nunique()," diffrent categories of crimes")
crime_dict = {}
for cat in merged_data["Category"].unique():
    crime_dict[cat] = merged_data[merged_data["Category"] == cat].shape[0]
    print(cat, ":", merged_data[merged_data["Category"] == cat].shape[0])
print("most common crime", max(crime_dict, key=crime_dict.get))
print("least common crime", min(crime_dict, key=crime_dict.get))
crime_dict = dict(sorted(crime_dict.items(), key=lambda item: item[1], reverse=True))

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 18))
axes[0].bar(crime_dict.keys(), crime_dict.values());
axes[0].set_xlabel("Crime Category")
axes[0].set_ylabel("Number of Crimes")
axes[0].tick_params(axis='x', rotation=30)

oldcrimes = dict(data_old["Category"].value_counts())
axes[1].bar(oldcrimes.keys(), oldcrimes.values());
axes[1].set_xlabel("Crime Category")
axes[1].set_ylabel("Number of Crimes")
axes[1].tick_params(axis='x', rotation=90)

newcrimes = dict(data_new["Incident Category"].value_counts())
axes[2].bar(newcrimes.keys(), newcrimes.values());
axes[2].set_xlabel("Crime Category")
axes[2].set_ylabel("Number of Crimes")
axes[2].tick_params(axis='x', rotation=90)
```

The plot above already gives us a bit of information but does not tell us anything about how crimes changed over the years. Let's have a look at it!

> *Exercise 3*: Temporal patterns.
> * To start off easily, let's count the number of crimes per year:
>     * What is the year with most crimes? 
>     * What is the year with the fewest crimes?.
> * Create a barplot of crimes-per-year (years on the $x$-axis, crime-counts on the $y$-axis).
> * Finally, Police chief Suneman is interested in the temporal development of only a **subset of categories, the so-called *focus crimes***. Those categories are listed below (for convenient copy-paste action). Create bar-charts displaying the year-by-year development of each of these categories across the years 2003-2025.
> 

```python
dict_crimes_year = {} 
for year in merged_data["Year"].unique():
    dict_crimes_year[year] = [((merged_data[merged_data["Year"] == year]["Category"] == cat).sum(),cat) for cat in merged_data["Category"].unique()]
print(dict_crimes_year)
print("year with most crimes", max(dict_crimes_year, key=dict_crimes_year.get))
print("year with least crimes", min(dict_crimes_year, key=dict_crimes_year.get))

fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(15, 25))
fig.suptitle("Number of Crimes per Year by Category", fontsize=24, y=0.92)
for i, (ax, crime) in enumerate(zip(axes.flatten(), merged_data["Category"].unique())):
    ax.bar(dict_crimes_year.keys(), [sum([x[0] for x in dict_crimes_year[year] if x[1] == crime]) for year in dict_crimes_year.keys()])
    ax.set_title(crime)
    ax.tick_params(axis='x', rotation=30)
    
    if i % 2 == 0:
        ax.set_ylabel("Number of Crimes")
    if i == len(axes.flatten()) - 2 or i == len(axes.flatten()) - 3:
        ax.set_xlabel("Year", fontsize=14)

axes.flatten()[-1].set_visible(False)
plt.tight_layout()
plt.subplots_adjust(top=0.9)
```
